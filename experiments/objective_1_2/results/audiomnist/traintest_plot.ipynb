```python
import json
import matplotlib.pyplot as plt

# === CONFIG ===
JSON_PATH = "training_results.json"   # Change path if needed
OUTPUT_PATH = "audio_training_curves.png"

# === LOAD TRAINING HISTORY ===
with open(JSON_PATH, "r") as f:
    history = json.load(f)

train_loss = history["training_history"]["train_loss"]
val_loss = history["training_history"]["val_loss"]
train_acc = history["training_history"]["train_acc"]
val_acc = history["training_history"]["val_acc"]

# === PLOT LOSS AND ACCURACY CURVES ===
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Plot Loss
ax[0].plot(train_loss, label='Train Loss', color='blue')
ax[0].plot(val_loss, label='Val Loss', color='orange')
ax[0].set_title("Training vs Validation Loss")
ax[0].set_xlabel("Epoch")
ax[0].set_ylabel("Loss")
ax[0].legend()
ax[0].grid(True)

# Plot Accuracy
ax[1].plot(train_acc, label='Train Accuracy', color='blue')
ax[1].plot(val_acc, label='Val Accuracy', color='orange')
ax[1].set_title("Training vs Validation Accuracy")
ax[1].set_xlabel("Epoch")
ax[1].set_ylabel("Accuracy")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.savefig(OUTPUT_PATH)
plt.show()
```
